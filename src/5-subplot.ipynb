### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Prepare data
x = np.linspace(0, 10, 100)
y = x ** 2

bar_categories = ['A', 'B', 'C', 'D']
bar_values = [5, 7, 3, 9]

scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]

pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Create 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Line chart
axs[0, 0].plot(x, y, color='blue')
axs[0, 0].set_title('Line Chart: y = x^2')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')
axs[0, 0].grid(True)

# Bar chart
axs[0, 1].bar(bar_categories, bar_values, color=['red', 'green', 'blue', 'orange'])
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Category')
axs[0, 1].set_ylabel('Value')
axs[0, 1].grid(axis='y', linestyle='--', alpha=0.7)

# Scatter plot
axs[1, 0].scatter(scatter_x, scatter_y, color='purple')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')
axs[1, 0].grid(True)

# Pie chart
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=140, colors=['gold', 'lightblue', 'lightgreen', 'salmon'])
axs[1, 1].set_title('Pie Chart')
axs[1, 1].axis('equal')  # Make it a circle

# Adjust layout and display
plt.tight_layout()
plt.show()
```
