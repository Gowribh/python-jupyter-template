### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
# Import necessary libraries
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'green', 'blue', 'orange']  # Different colors for each bar

# Create the bar plot
plt.figure(figsize=(8, 5))
plt.bar(categories, values, color=colors)

# Add title and labels
plt.title('Bar Plot of Categories')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show the plot
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Optional: add horizontal gridlines
plt.tight_layout()
plt.show()
```
